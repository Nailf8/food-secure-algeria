```python
from llama_index.llms.databricks import Databricks
from langchain_huggingface import HuggingFaceEmbeddings
from llama_index.embeddings.langchain import LangchainEmbedding
from llama_index.core import Settings


Settings.llm = Databricks(
    model="databricks-meta-llama-3-1-70b-instruct",
    api_key=os.environ["DATABRICKS_TOKEN"],
    api_base="https://adb-7215147325717155.15.azuredatabricks.net/serving-endpoints",
)
Settings.embed_model = LangchainEmbedding(HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2"))
```

```python
from llama_index.core import Document, VectorStoreIndex

# Convert in llama index document format
documents = [Document(text=t) for t in splits]
```

```python
from llama_index.core.node_parser import SentenceSplitter

node_parser = SentenceSplitter(chunk_size=30000)

nodes = node_parser.get_nodes_from_documents(documents)
```

```python
from pinecone import Pinecone, ServerlessSpec

load_dotenv()

pc = Pinecone(api_key=os.environ["PINECONE_API_KEY"])

pc.create_index(
    name="sidindex",
    dimension=384,
    metric="dotproduct",
    spec=ServerlessSpec(cloud="aws", region="us-east-1"),
)


pinecone_index = pc.Index("sidindex")
```

```python
from llama_index.vector_stores.pinecone import PineconeVectorStore
from llama_index.core import StorageContext, VectorStoreIndex


vector_store = PineconeVectorStore(
    pinecone_index=pinecone_index,
    add_sparse_vector=True,
)
storage_context = StorageContext.from_defaults(vector_store=vector_store)
index = VectorStoreIndex.from_documents(
    documents, storage_context=storage_context
)
```

```python
from llama_index.vector_stores.pinecone import PineconeVectorStore
from llama_index.core import VectorStoreIndex

pc = Pinecone(api_key=os.environ["PINECONE_API_KEY"])
pinecone_index = pc.Index('sidindex')

vector_store = PineconeVectorStore(
    pinecone_index=pinecone_index, text_key="text"
)
retriever = VectorStoreIndex.from_vector_store(vector_store).as_retriever(
    similarity_top_k=10
)
query_engine = VectorStoreIndex.from_vector_store(vector_store).as_query_engine(
    similarity_top_k=10
)
```

```python
query = "Pourcentage des terres arables aménagées pour l'irrigation (%) (moyenne sur 3 ans) "
nodes = retriever.retrieve(query)
response_without_rerank = query_engine.query(query)
```

```python
from llama_index.core.response.pprint_utils import pprint_response
from llama_index.core.response.notebook_utils import display_source_node

pprint_response(response_without_rerank, show_source=True)

for node in nodes: 
    display_source_node(node)
```

```python
import os
from llama_index.postprocessor.cohere_rerank import CohereRerank


api_key = os.environ["COHERE_API_KEY"]
cohere_rerank = CohereRerank(api_key=api_key, top_n=5)
```

```python
engine = VectorStoreIndex.from_vector_store(vector_store).as_query_engine(
    vector_store_query_mode="hybrid", 
    similarity_top_k=10,
    node_postprocessors=[cohere_rerank], 
    alpha=0.0
)
```

```python
response = engine.query("Le pourcentage de terres arables aménagées pour l'irrigation")
```

```python
from llama_index.core.response.pprint_utils import pprint_response

pprint_response(response, show_source=True)
```
